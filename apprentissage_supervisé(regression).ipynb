**Visualiser le nuage de point**

```python
import matplotlib.pyplot as plt

plt.scatter(df['area'], df['price'], color='blue')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

**Installons Scikit-learn**

L'ensemble des algorithmes de machine learning se trouve en majeur partie dans cette bibliothèque (Régression, classification)

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = df[['area']]
y = df['price']
model.fit(x, y)
```

```python
price = model.predict([[120]])
print(price)
```

Evaluation du model de prédiction

```python
from sklearn.metrics import r2_score
y_pred = model.predict(x)
r2 = r2_score(y, y_pred)
print(f"La valeur de l'évaluation du model est  : {r2*100} %")
```
